```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

pd.set_option('max_columns', 100)
%matplotlib inline
```

```python
df_train = pd.read_csv('./df_train_for_fit.csv')
df_test = pd.read_csv('./df_test_for_fit.csv')
df = pd.concat([df_train, df_test], axis=0)
```

```python
df.columns
```

```python
X = df_train.drop(['y'], axis=1).values
y = df_train['y'].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.3, random_state=0)
```

## 性能評価

### random forest

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=100)
```

```python
rfc.fit(X_train, y_train)
```

#### feature importance

```python
feature_names = df.drop(['y'], axis=1).columns
n_features = len(feature_names)
plt.figure(figsize=(10, 8))
plt.barh(range(n_features), rfc.feature_importances_ , align='center')
plt.yticks(np.arange(n_features), feature_names)
plt.show()
```

```python
feature_Ser = pd.Series(rfc.feature_importances_, index=feature_names)
feature_Ser = feature_Ser.sort_values(ascending=False)
feature_Ser.index[:20]
```

#### スコア

```python
y_train_pred_rfc = rfc.predict_proba(X_train)
y_test_pred_rfc = rfc.predict_proba(X_test)

from sklearn.metrics import roc_auc_score
print(roc_auc_score(y_train, y_train_pred_rfc[:,1]))
print(roc_auc_score(y_test, y_test_pred_rfc[:,1]))
```

### 勾配Boosting木

```python
from sklearn.ensemble import GradientBoostingClassifier
gbc = GradientBoostingClassifier()
```

```python
gbc.fit(X_train, y_train)
```

```python
feature_names = df.drop(['y'], axis=1).columns
n_features = len(feature_names)
plt.figure(figsize=(10, 8))
plt.barh(range(n_features), gbc.feature_importances_  , align='center')
plt.yticks(np.arange(n_features), feature_names)
plt.show()
```

#### スコア

```python
y_train_pred_gbc = gbc.predict_proba(X_train)
y_test_pred_gbc = gbc.predict_proba(X_test)

from sklearn.metrics import roc_auc_score
print(roc_auc_score(y_train, y_train_pred_gbc[:,1]))
print(roc_auc_score(y_test, y_test_pred_gbc[:,1]))
```

### 勾配Boosting木 + Grid Search

```python
from sklearn.ensemble import GradientBoostingClassifier
gbc = GradientBoostingClassifier()
from sklearn.model_selection import GridSearchCV
param_grid = {
    'n_estimators' : [100],
    'learning_rate': [0.1, 0.05, 0.01],
    'max_depth': [2, 3, 4],
    #'min_samples_leaf': [1, 10, 50],
    #'max_features': [1.0, 0.3, 0.1] 
              }
gs = GridSearchCV(gbc, param_grid=param_grid, cv=4, verbose=2)
```

```python
#gs.fit(X_train, y_train)
```

```python
feature_names = df.drop(['y'], axis=1).columns
n_features = len(feature_names)
plt.figure(figsize=(10, 8))
plt.barh(range(n_features), gs.best_estimator_.feature_importances_ , align='center')
plt.yticks(np.arange(n_features), feature_names)
plt.show()
```

#### スコア

```python
y_train_pred_gs = gs.best_estimator_.predict_proba(X_train)
y_test_pred_gs = gs.best_estimator_.predict_proba(X_test)
```

```python
from sklearn.metrics import roc_auc_score
print(roc_auc_score(y_train, y_train_pred_gs[:,1]))
print(roc_auc_score(y_test, y_test_pred_gs[:,1]))
```

### ロジスティック回帰

```python
from sklearn.linear_model import LogisticRegressionCV
lrcv = LogisticRegressionCV(Cs=[0.001, 0.01, 0.1, 1.0])
lrcv.fit(X_train, y_train)
```

#### スコア

```python
from sklearn.metrics import roc_auc_score
print(roc_auc_score(y_train, y_train_pred_lrcv[:,1]))
print(roc_auc_score(y_test, y_test_pred_lrcv[:,1]))
```
